```python
with open("app.py", "w") as f:
    f.write("""
import streamlit as st
from PIL import Image
import google.generativeai as genai
import os
import random
import io

# Configure Gemini API
genai.configure(api_key="AIzaSyAoIYOuO_2Yz6LmPhQj0q8T2HRROIMrNy4")
model = genai.GenerativeModel("gemini-1.5-pro")

def generate_text_response(prompt):
    response = model.generate_content(prompt)
    return response.text

def process_image(image, text_prompt=""):
    img_byte_arr = io.BytesIO()
    image.save(img_byte_arr, format='JPEG')
    img_byte_arr = img_byte_arr.getvalue()
    response = model.generate_content([text_prompt, {"mime_type": "image/jpeg", "data": img_byte_arr}])
    return response.text

def process_text_image(image, text_prompt):
    img_byte_arr = io.BytesIO()
    image.save(img_byte_arr, format='JPEG')
    img_byte_arr = img_byte_arr.getvalue()
    response = model.generate_content([text_prompt, {"mime_type": "image/jpeg", "data": img_byte_arr}])
    return response.text

# Streamlit UI
st.title("Gemini LLM Multi-Modal AI Hands-On")
st.subheader("Bring Your Own Dataset (Kaggle, Local, APIs)")

# Text Processing
st.write("## Text Processing")
text_input = st.text_area("Enter a prompt:")
if st.button("Generate Text Response"):
    if text_input:
        result = generate_text_response(text_input)
        st.write("Generated Response:", result)
    else:
        st.warning("Please enter a prompt.")

# Image Processing
st.write("## Image Processing")
uploaded_file = st.file_uploader("Upload an Image", type=["jpg", "png", "jpeg"])
if uploaded_file:
    st.image(uploaded_file, caption="Uploaded Image", use_column_width=True)
    if st.button("Analyze Image"):
        image = Image.open(uploaded_file)
        result = process_image(image)
        st.write("Image Analysis Result:", result)

# Text + Image Processing
st.write("## Text + Image Processing")
text_prompt_2 = st.text_area("Enter a text prompt for the image:")

# NEW: File uploader specifically for this section
uploaded_file_2 = st.file_uploader("Upload an Image for Text + Image Processing", type=["jpg", "png", "jpeg"], key="text_image_upload")

if uploaded_file_2 and text_prompt_2:
    if st.button("Analyze Text + Image"):
        image = Image.open(uploaded_file_2)
        result = process_text_image(image, text_prompt_2)
        st.image(uploaded_file_2, caption="Uploaded Image for Text + Image Processing", use_column_width=True)
        st.write("Text + Image Analysis Result:", result)
elif st.button("Analyze Text + Image"):
    st.warning("Please upload an image and enter a text prompt.")

# Kaggle Dataset Integration
st.write("## Load Kaggle Dataset")
dataset_name = st.text_input("Enter Kaggle dataset name:")
if st.button("Download Dataset"):
    if dataset_name:
        os.system(f"kaggle datasets download -d {dataset_name} -p ./data")
        os.system(f"unzip -o './data/{dataset_name}.zip' -d ./data")

        # Extract nested zip files if present
        for root, dirs, files in os.walk("./data"):
            for file in files:
                if file.endswith(".zip"):
                    zip_path = os.path.join(root, file)
                    os.system(f"unzip -o '{zip_path}' -d './data'")

        st.success(f"Downloaded {dataset_name} successfully!")

        # Random Image Processing from Dataset
        st.write("### Random Image Analysis")
        image_extensions = [".jpg", ".jpeg", ".png"]
        image_files = [os.path.join(root, file)
                       for root, _, files in os.walk("./data")
                       for file in files if file.lower().endswith(tuple(image_extensions))]

        if image_files:
            random_image_path = random.choice(image_files)
            st.image(random_image_path, caption="Random Image from Dataset", use_column_width=True)

            random_text_prompt = "Analyze this image and describe its content."
            result = process_text_image(Image.open(random_image_path), random_text_prompt)
            st.write("Random Image Analysis Result:", result)
        else:
            st.warning("No images found in the downloaded dataset.")
    else:
        st.warning("Please enter a Kaggle dataset name.")
""")
```
